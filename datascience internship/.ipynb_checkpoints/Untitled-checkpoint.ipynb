```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'g')
plt.show()
```

```python
plt.plot(x,y,color='0')
```

```python
y=[i**2 for i in x]
plt.plot(x,y,c='y')
plt.title("graph of ")
plt.xlabel("")
plt.ylabel("power of x values")
plt.show()
```

```python
import numpy as np
x=np.linspace()
```

```python
plt.plot(x,y,'#000000',linestyle='-.',linewidth=1,marker='p',label="line1")
plt.plot(x+2,y,'#0000ff',linestyle='--',marker='p',labe1='line2')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("multiple plots")
plt.legend(loc='best')
plt.show()
```

```python
fig,a=plt.subplots(2,2)
a[0][0].plot(x,x*x)
a[0][0].set_title('square')
a[0][1].plot(x,np.sqrt(x))
a[0][1].set_title('square root')
a[1][0].plot(x,np.exp(x))
a[1][0].set_title('exponential')
a[1][1].plot(x,np.log(x))
a[1][1].set_title('logarithm')
plt.show()
```
